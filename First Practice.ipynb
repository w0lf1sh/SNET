##  <font color=blue>  Trends into a pandas data frame </font>
Data frame will be the main data structure (in this subject) for analysing the data from twitter.

```python
import pandas as pd
trends_df = pd.DataFrame.from_dict(top_trends_world[0]['trends'])
trends_df
```

### Structure of the data frame

```python
trends_df.columns
```

### Sort by 'tweet_volume', reorder columns

```python
trends_df.sort_values(by = ['tweet_volume'], ascending=False)[['tweet_volume','name','query']]
```

### Lets add a new column and sort by several criteria

```python
trends_df['Length name'] = trends_df['name'].apply(lambda x: len(x))
trends_df[['tweet_volume','name','Length name']]
```

### sort by several criterion, and save in place.

```python
trends_df.sort_values(by = ['tweet_volume','Length name'], ascending = False, inplace = True)
trends_df[['tweet_volume','Length name']]
```
