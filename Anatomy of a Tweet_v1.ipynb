## <font color='red'>Homework:</font>

* Create a dataframe with a column with the identity of the user and a column with the identity of the user mentioned in the tweet.
* Use (Future):  creating an graph with an implicit relation.
* Finally order the DataFrame by the number of followers of the user.

```python
identities_df = pd.DataFrame(columns=['user', 'mentioned_user', 'followers'])
```

```python
identities_df['user'] = DataBase_df['user.name']
```

```python
identities_df
```

```python
mentioned_users = []

for i in range(0, len(identities_df)):
    if DataBase_df['entities.user_mentions'][i]:
        mentioned_users.append(DataBase_df['entities.user_mentions'][i][0]['name'])
    else:
        mentioned_users.append('NaN')
```

```python
identities_df['mentioned_user'] = pd.Series(mentioned_users)
identities_df
```

```python
identities_df['followers'] = DataBase_df['user.followers_count']
identities_df.sort_values(by=['followers'], ascending=False)
```
