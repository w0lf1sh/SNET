```python
ListEntities = ['hashtags',  'user_mentions']
def ListEntity(Entity = 'hashtags',Column = 'text'):
    List = list()
    for row in EntitiesFlatten[Entity].iteritems():
        if len(row[1])>0:
            List.append(row[1][0][Column])
        else:
            List.append('')
    return List
Hashtags = ListEntity(Entity = 'hashtags')    
User_mentions = ListEntity(Entity = 'user_mentions',Column = 'name')   
print(f'Check lengths of extracted info: \n \tHashtags:{len(Hashtags)},User_mentions:{len(User_mentions)}')
```

### Lets introduce a usefull feature: "Counter"
We could like to know the frequency of entities

```python
from collections import Counter

def InfoEntities(List = User_mentions):
    CountEntity = Counter(List)
    CountEntity.pop('',None)
    CountEntity = dict(CountEntity.most_common(15))
    print(CountEntity)
    plt.bar(CountEntity.keys(), CountEntity.values())
    plt.title('15 most common entities')
    plt.xticks(rotation='vertical')
    plt.show()
    return None

InfoEntities(List = Hashtags)
InfoEntities(List = User_mentions)
```

# Graph creation and analysis tool: networkx
https://networkx.org/

```python
g= nx.DiGraph()
for row in DataBase_user.iterrows():
    info = row[1]
    if len(info['Hashtags']) >0 and len(info['User_mentions'])>0:
        g.add_node(info['Hashtags'])
        g.add_node(info['User_mentions'])
        g.add_edge(info['Hashtags'], info['User_mentions'])
        g.add_edge(info['User_mentions'], info['Hashtags'])
print(f'Total nodes:{len(g.nodes)}')
```

### <font color='blue'> Answer: </font>
In this adjacency matrix, similar to the H matrix we have seen in class, we can see the structure of the network. Nodes are designed by hashtags pointing to User mentions. Following this idea, with the rows we can see which hashtags point to User mentions (denoted by 1) and with the columns we can check which User mentions are pointed by hashtags.<br>
Since this matrix represents the structure of the network, we can have dangling nodes and especially isolated communities which can not be reached. To solve those problems we could normalize the rows to get a stochastic matrix, so that all the rows add to 1, and also if there was a dangling node with no outputs, instead of being all zeroes, the row could add up to 1 with equally distributed values.<br>
To attack the problem of isolated components, we could add a damping factor together with a teleportation matrix. This way a random surfer that might be surfing the graph may jump from one isolated community to another with a probability corresponding to the damping factor.
