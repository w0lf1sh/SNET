```python
EntitiesFlatten = pd.DataFrame.from_dict(pd.json_normalize(DataBase_df['entities']))
EntitiesFlatten.index = DataBase_df.index
DataBase_df2 = pd.concat([DataBase_df,EntitiesFlatten],axis = 1)
FilteredDataBase_df = DataBase_df[DataBase_df['entities'].isnull() == False]
EntitiesFlatten = pd.DataFrame.from_dict(pd.json_normalize(FilteredDataBase_df['entities']))
EntitiesFlatten.columns
```

```python
ListEntities = ['hashtags',  'user_mentions']
def ListEntity(Entity = 'hashtags',Column = 'text'):
    List = list()
    for row in EntitiesFlatten[Entity].iteritems():
        if len(row[1])>0:
            List.append(row[1][0][Column])
        else:
            List.append('')
    return List
Hashtags = ListEntity(Entity = 'hashtags')    
User_mentions = ListEntity(Entity = 'user_mentions',Column = 'name')   
print(f'Check lengths of extracted info: \n \tHashtags:{len(Hashtags)},User_mentions:{len(User_mentions)}')
```

### Lets introduce a usefull feature: "Counter"
We could like to know the frequency of entities

```python
from collections import Counter

def InfoEntities(List = User_mentions):
    CountEntity = Counter(List)
    CountEntity.pop('',None)
    CountEntity = dict(CountEntity.most_common(15))
    print(CountEntity)
    plt.bar(CountEntity.keys(), CountEntity.values())
    plt.title('15 most common entities')
    plt.xticks(rotation='vertical')
    plt.show()
    return None

InfoEntities(List = Hashtags)
InfoEntities(List = User_mentions)
```

```python
Columns_with_user = ['name','full_text','location','followers_count','friends_count','Hashtags',  'User_mentions','Lang']
DataBase_user = DataBase_df3[Columns_with_user]
DataBase_user.head().T
```

# Graph creation and analysis tool: networkx
https://networkx.org/

```python
g= nx.DiGraph()
for row in DataBase_user.iterrows():
    info = row[1]
    if len(info['Hashtags']) >0 and len(info['User_mentions'])>0:
        g.add_node(info['Hashtags'])
        g.add_node(info['User_mentions'])
        g.add_edge(info['Hashtags'], info['User_mentions'])
        g.add_edge(info['User_mentions'], info['Hashtags'])
print(f'Total nodes:{len(g.nodes)}')
```

## The objective is just to plot a graph.
### In the future we will explore the specific bipartite plot.
Now we will try several plots.

```python
plt.figure(figsize=(15,15)) 
nx.draw(g, with_labels=True,alpha = 0.75,font_size =12,pos=nx.spring_layout(g))
plt.show()
```

```python
plt.figure(figsize=(15,15)) 
nx.draw(g, with_labels=True,alpha = 0.75,font_size =12,pos=nx.planar_layout(g))
plt.show()
```

```python
plt.figure(figsize=(15,15)) 
nx.draw(g, with_labels=True,alpha = 0.75,font_size =12,pos=nx.circular_layout(g))   
plt.show()
```

## <font color='red'> Homework: </font>
From the above relationships what would you conclude?, if you find an interesting pattern, look into the dataframe and see the texts associated either with the interesting hashtag or mention.

See next section for a selection of relationships for the nodes with most followers.

```python
g= nx.DiGraph()
for row in DataBase_user.iterrows():
    info = row[1]
    if len(info['Hashtags']) >0 and len(info['User_mentions'])>0:
        if info['followers_count'] >5000:
            g.add_node(info['Hashtags'])
            g.add_node(info['User_mentions'])
            g.add_edge(info['Hashtags'], info['User_mentions'])
print(f'Total nodes:{len(g.nodes)}')
```

## Lets see the structure of the nodes with more followers.

```python
nx.draw(g, with_labels=True,alpha = 0.75,font_size =8,pos=nx.spring_layout(g))
```
